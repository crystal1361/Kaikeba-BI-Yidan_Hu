```python
from sklearn.datasets import load_boston
```

```python
#数据加载
boston=load_boston()
```

```python
X=boston['data']
y=boston['target']
```

```python
print(X)
```

```python
print(X.shape)
```

```python
print(y)
```

```python
#转变y的形状
y=y.reshape(-1,1)
```

```python
print(y)
```

```python
#对数据进行规范化
from sklearn.preprocessing import MinMaxScaler
ss_input=MinMaxScaler()
print(X)
type(X)
```

```python
X=ss_input.fit_transform(X)
print(X)
```

```python
#把numpy放进tensor里再切分
X=torch.from_numpy(X).type(torch.FloatTensor)
y=torch.from_numpy(y).type(torch.FloatTensor)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X,y,test_size=0.25)
```

```python
#构造网络
from torch import nn
model=nn.Sequential(
    nn.Linear(13,16),       #13个特征，16个神经元
    nn.ReLU(),
    nn.Linear(16,1)  # 输出只有一个    
)
```

```python
#定义优化器和损失函数
criterion=nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01) #learningrate一般是0.01或0.001，torch里面很多优化所以学习速度可以高一点
```

```python
#训练
max_epoch=300
iter_loss=[]
for i in range(max_epoch):
    #前向传播
    y_pred=model(X)
    #计算loss
    loss=criterion(y_pred, y)
    iter_loss.append(loss.item())
    print(loss.item())
    #清空之前的梯度
    optimizer.zero_grad()
    #反向传播
    loss.backward()
    #参数更新
    optimizer.step()
```

```python
print(loss.item())
```

```python
#测试
output=model(test_x)
predict_list=output.detach().numpy() #把预测结果放到numpy里来取
print(predict_list)
```

```python
# 绘制不同iteration的loss
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(max_epoch)
y=np.array(iter_loss)
plt.plot(x,y)
plt.title('Loss Value in 300 epoch')
plt.xlabel('Epoch')
plt.ylabel('Mean Loss Value')
```

```python
print(y)
```

```python
#做散点图，查看真实值与预测值的相关程度
x=np.arange(test_x.shape[0])
y1=np.array(predict_list)
y2=np.array(test_y)
line1=plt.scatter(x,y1,c='red')
line2=plt.scatter(x,y2,c='yellow')
plt.legend([line1,line2],['predict','real'])
plt.title('Prediction VS Real')
plt.ylabel('House Price')
```
