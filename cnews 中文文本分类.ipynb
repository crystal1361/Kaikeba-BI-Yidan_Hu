```python
import torch
from torch import nn
from torch import optim
from model import TextRNN
from cnews_loader import read_vocab, read_category, process_file
```

```python
# 设置数据目标
train_file = 'cnews.train.txt'
test_file = 'cnews.test.txt'
val_file = 'cnews.val.txt'
vocab_file = 'cnews.vocab.txt'
```

```python
import numpy as np
def train():
    #创建模型
    model = TextRNN()
    #设置损失函数
    Loss = nn.MultiLabelSoftMarginLoss()
    optimizer=optim.Adam(model.parameters(), lr=0.001)
    best_val_acc=0
    for epoch in range(1000):
        print('epoch=', epoch)
        
        for step, (x_batch, y_batch) in enumerate(train_loader):
            x = x_batch
            y = y_batch
            out=model(x)
            loss=Loss(out, y)
            print('loss=', loss)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            accuracy = np.mean((torch.argmax(out, 1)==torch.argmax(y, 1)).numpy())
            print(accuracy)
        # 对模型进行验证
        if (epoch+1) % 5 == 0:
            for step, (x_batch, y_batch) in enumerate(val_loader):
                x = x_batch
                y = y_batch
                out = model(x)
                accuracy = np.mean((torch.argmax(out, 1)==torch.argmax(y, 1)).numpy())
                if accuracy>best_val_acc:
                    torch.save(model.state_dict(), 'model_best.pkl')
                    best_val_acc = accuracy
                    print('model_best.pkl update')
                    print(accuracy)
```

```python
# 获取新闻的类别和对应的id的字典
categories, cat_to_id = read_category()
# 获取新闻中所有出现过的字以及对应的id的字典
words, word_to_id = read_vocab(vocab_file)
vocab_size = len(words)
vocab_size
```

```python
x_train, y_train = process_file(train_file, word_to_id, cat_to_id, max_length=600)
x_val, y_val = process_file(val_file, word_to_id, cat_to_id, max_length=600)
```

```python
# 设置使用GPU
# cuda = torch.device('cuda')
x_train, y_train = torch.LongTensor(x_train), torch.LongTensor(y_train)
x_val, y_val = torch.LongTensor(x_val), torch.LongTensor(y_val)
```

```python
import torch.utils.data as Data
train_dataset = Data.TensorDataset(x_train, y_train)
train_loader = Data.DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)

val_dataset=Data.TensorDataset(x_val, y_val)
val_loader = Data.DataLoader(dataset=val_dataset, batch_size=64, shuffle=True)
```

```python
train()
```
