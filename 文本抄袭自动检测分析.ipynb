```python
#加载停用词
with open('chinese_stopwords.txt', 'r', encoding='utf-8') as file:
    stopword = [i[:-1] for i in file.readlines()]
print(stopword)
```

```python
#数据加载
import pandas as pd
news = pd.read_csv('sqlResult.csv', encoding='gb18030')
news.head()
```

```python
news.shape
```

```python
#处理缺失值
news=news.dropna(subset=['content'])
```

```python
news.shape
```

```python
#分词
import jieba
def split_text(text):
    text = text.replace(' ','').replace('\n','')
    text2 = jieba.cut(text.strip())
    #去掉停用词
    result = ' '.join([w for w in text2 if w not in stopword])
    return result
print(news.iloc[0])
```

```python
print(news.iloc[0].content)
```

```python
print(split_text(news.iloc[0].content))
```

```python
#对所有文本进行分词
corpus = list(map(split_text, [str(i) for i in news.content]))
```

```python
print(corpus[0])
```

```python
print(corpus[0:2])
```

```python
import pickle, os
if not os.path.exists('corpus.pkl'):
    #把corpus保存到文件file中去，方便下次使用
    with open('corpus.pkl','wb') as file: 
        pickle.dump(corpus, file)
else:
    #调用上次处理的结果
    with open('corpus.pkl', 'rb') as file:
        corpus = pickle.load(file)
        
#计算corpus中的TFIDF矩阵
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
countvectorizer = CountVectorizer(encoding='gb18030', min_df=0.015)
tfidftransformer = TfidfTransformer()
#先做TF，再做IDF =>TFIDF
countvector=countvectorizer.fit_transform(corpus) #fit是学习数据中的参数，transform是根据学到的参数，进行预处理，得到想要的结果
print(countvector) #np.array的简化写法。第0行第174列的词出现了一次。
print(countvector.toarray()) #corpus每句话是一行，第174列的词在第0行这句话中出现了一次。
print(countvector.toarray()[0,174])
```

```python
tfidf = tfidftransformer.fit_transform(countvector)
print(tfidf)
```

```python
type(tfidf)
```

```python
#标记是否为自己的新闻
label = list(map(lambda source:1 if '新华' in str(source) else 0,news.source))
print(label)
```

```python
#数据切分
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf.toarray(), label, test_size=0.3) # Allowed inputs are lists, numpy arrays, scipy-sparse matrices or pandas dataframes.
```

```python
from sklearn.naive_bayes import MultinomialNB 
clf = MultinomialNB()
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
print(y_predict)
```

```python
#使用模型检测抄袭新闻，预测风格 （看预测的label是否为新华社）
#predition代表 风格预测， prediction=1 代表风格预测为新华社
#labels代表实际值，如果label=1 说明实际为新华社
prediction = clf.predict(tfidf.toarray())
```

```python
import numpy as np
labels = np.array(label) #实际的labels
```

```python
#compare_news_index有两列，prediction为预测风格，labels为实际
compare_news_index = pd.DataFrame({'prediction': prediction, 'labels': labels})
print(compare_news_index)
```

```python
#有嫌疑的index：预测为新华社风格，但label又不是新华社
copy_news_index = compare_news_index[(compare_news_index['prediction'] == 1) & (compare_news_index['labels'] == 0)]
print(copy_news_index)
```

```python
print(copy_news_index.index)
```

```python
#有嫌疑的文章的index
potentialindex=copy_news_index.index
```

```python
#实际为新华社的新闻
xinhuashe_news_index = compare_news_index[(compare_news_index['labels'] == 1)].index
print(xinhuashe_news_index)
```

```python
print('可能为copy的新闻条数', len(copy_news_index))
```

```python
#使用kmeans对文章进行聚类
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
normalizer = Normalizer()
scaled_array = normalizer.fit_transform(tfidf.toarray())
```

```python
# 25是人工定义的桶的个数，聚类的个数
kmeans = KMeans(n_clusters=25)
k_labels = kmeans.fit_predict(scaled_array)
```

```python
print(k_labels)
```

```python
# 创建id_class
id_class = {index:class_ for index, class_ in enumerate(k_labels)}
print(id_class)
```

```python
from collections import defaultdict
class_id = defaultdict(set)
for index, class_ in id_class.items(): #Python Dictionary items() Method returns key-value pairs of the dictionary, as tuples in a list.
    # 只统计新华社发布的class_id
    if index in xinhuashe_news_index.tolist():
        class_id[class_].add(index)
```

```python
print(class_id)
```

```python
print(class_id.keys())
```

```python
# 查找相似文本
from sklearn.metrics.pairwise import cosine_similarity
def find_similar_text(cpindex, top=10):
    #只在新华社发布的文章中查找
    dist_dict = {i:cosine_similarity(tfidf[cpindex], tfidf[i]) for i in class_id[id_class[cpindex]]}
    #从大到小进行排序
    return sorted(dist_dict.items(), key=lambda x:x[1][0], reverse=True)[:top]
```

```python
# 测试有嫌疑的文章与哪些新华社的文章相似
cpindex=potentialindex[2]
print(cpindex)
similar_list = find_similar_text(cpindex)
print(similar_list)
```
